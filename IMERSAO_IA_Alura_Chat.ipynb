```python
# CÃ©lula 3: Imports e nova funÃ§Ã£o para rodar mensagem (TENTATIVA 6)

# Importa as classes e funÃ§Ãµes necessÃ¡rias
from google.adk.agents import Agent
from google.adk.runners import Runner # Runner continua necessÃ¡rio para criar a instÃ¢ncia
from google.adk.sessions import InMemorySessionService # InMemorySessionService tb
from google.adk.tools import google_search # <-- VERIFIQUE Google Search
from google.genai import types # Importar 'types' Ã© importante!

from datetime import date
import textwrap
from IPython.display import display, Markdown, HTML
import requests
import warnings

warnings.filterwarnings("ignore")

# --- REMOVIDA A FUNÃ‡ÃƒO call_agent ---
# --- Nova funÃ§Ã£o para rodar uma Ãºnica mensagem atravÃ©s de um Runner existente ---
# Esta funÃ§Ã£o recebe uma instÃ¢ncia de Runner e o texto da mensagem.
# O Runner passado deve ter sido criado com o SessionService e Agent corretos.
def run_agent_message(runner: Runner, message_text: str) -> str:
    """
    Envia uma mensagem de texto para um Runner EXISTENTE e retorna a resposta final.
    Assume user_id='user1' e session_id='session1' para a interaÃ§Ã£o via Runner.
    """
    # Cria o conteÃºdo da mensagem de entrada no formato esperado pelo ADK/GenAI
    content = types.Content(role="user", parts=[types.Part(text=message_text)])

    final_response = ""
    try:
        # Usa o Runner que foi passado como argumento.
        # O Runner.run() ainda precisa do user_id e session_id mesmo tendo o SessionService,
        # para identificar a sessÃ£o especÃ­fica DENTRO do SessionService.
        # user_id e session_id fixos aqui.
        for event in runner.run(user_id="user1", session_id="session1", new_message=content):
             if event.is_final_response():
                for part in event.content.parts:
                    if part.text is not None:
                        final_response += part.text
                        if len(event.content.parts) > 1 and part != event.content.parts[-1]:
                             final_response += "\n"
    except Exception as e:
        print(f"Ocorreu um erro durante a execuÃ§Ã£o do Runner: {e}")
        final_response = f"Desculpe, encontrei um problema ao processar sua solicitaÃ§Ã£o ({type(e).__name__}). Por favor, tente novamente."
        # Para debug, vocÃª pode descomentar as linhas abaixo:
        # import traceback
        # traceback.print_exc()

    return final_response.strip() # Remove espaÃ§os em branco no inÃ­cio/fim
```

```python
# FunÃ§Ã£o auxiliar para exibir texto formatado em Markdown no Colab
def to_markdown(text):
  """Converte texto para formato Markdown com identaÃ§Ã£o para exibiÃ§Ã£o."""
  # Substitui marcadores de lista comuns por marcadores markdown identados
  text = text.replace('â€¢', '  *') # Substitui bullet points por * com indentaÃ§Ã£o
  # Outras substituiÃ§Ãµes de formataÃ§Ã£o podem ser adicionadas aqui, se necessÃ¡rio
  lines = text.split('\n')
  formatted_lines = [f'> {line}' if line.strip() else '>' for line in lines] # Adiciona "> " no inÃ­cio de cada linha para visualizaÃ§Ã£o como citaÃ§Ã£o/bloco
  return Markdown('\n'.join(formatted_lines))
```
